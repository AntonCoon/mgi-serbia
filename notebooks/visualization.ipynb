```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas
import tqdm.notebook as tqdm
```

```python
data = pandas.read_csv("../data/final_clusterised_data.csv", index_col=0)
```

```python
min_x = min(data.x)
min_y = min(data.y)
coordinates = []
colors = []
for _, row in data.iterrows():
    _, x, y, _, color = tuple(row.values)
    coordinates.append((x - min_x, y - min_y))
    colors.append(color)
```

```python
max_x = max(data.x) - min_x + 1
max_y = max(data.y) - min_y + 1
radius = 28
offset = radius // 2 + 4
```

```python
canvas = np.zeros((max_x + radius + offset, max_y + radius + offset, 3))
count = np.ones((max_x + radius + offset, max_y + radius + offset))
```

```python
neighborhood = []
for dx in range(-radius, radius + 1):
    for dy in range(-radius, radius + 1):
        if dx ** 2 + dy ** 2 <= radius ** 2:
            neighborhood.append((dx, dy))
```

```python
for coord, color in tqdm.tqdm(zip(coordinates, colors)):
    x, y = coord
    canvas[x + offset, y + offset] += matplotlib.colors.to_rgb(color)
    count[x + offset, y + offset] += 1
    for dx, dy in neighborhood:
        canvas[x + offset + dx, y + offset + dy] += matplotlib.colors.to_rgb(color)
        count[x + offset + dx, y + offset + dy] += 1 
```

```python
mean_canvas = np.divide(
    canvas, 
    count[..., np.newaxis], 
    out=np.zeros_like(canvas), 
    where=(count[..., np.newaxis] != 0)
)
```

```python
plt.figure(figsize=(5490 // 300, 2824 // 300))
plt.imshow(mean_canvas)
plt.axis('off')
plt.savefig("canvas.png")
plt.show()
```
