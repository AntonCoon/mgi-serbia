```python
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import networkx
import numpy
import pandas
import seaborn
import scipy
import tqdm.notebook as tqdm
import time

numpy.seterr(divide='ignore', invalid='ignore')
seaborn.set_style("whitegrid")
```

# Clusterization of gene expression graph.
Let's define expression graph as weighted undirected graph where each node is a cell from tissue and edge u, v exist if u, v cells gene expressions are similar, and besides weight of (u, v) proportional to expression similarity and real(euclidean) distance between cells.

### Upload the cells data with precalqulated gene expression coefficients

```python
normalized_cell_data = pandas.read_csv("../data/filtered_normalized_expression_in_cells.csv", index_col=0)
```

```python
total_cell_amount, total_gene_amount = normalized_cell_data.shape
total_gene_amount -= 3
print(f"In the dataset we have {total_cell_amount} cells and {total_gene_amount} genes.")
```

Cells order and genes order will remain the same during all further computations.

```python
ordered_cell_ids = list(normalized_cell_data.cell_id)
ordered_gene_names = list(normalized_cell_data.columns[3:])
```

```python
cells_coordinates = numpy.array(normalized_cell_data[["x", "y"]])
cells_gene_expression = numpy.array(normalized_cell_data[ordered_gene_names])
```

Upload precalculated probabilities of gene expression graph edges which are based on distance between cells and their gene ecpression.

```python
with open("../data/dist_probs.npy", "rb") as f:
    array_of_all_distance_probs = numpy.load(f)
with open("../data/gene_probs.npy", "rb") as f:
    array_of_all_gene_exp_probs = numpy.load(f)
```

```python
real_cell_poses = {idx: (coord[0], coord[1]) for idx, coord in enumerate(cells_coordinates)}
```

### Build gene expression graph

```python
weighted_cells_graph = networkx.Graph()
for idx in range(len(ordered_cell_ids)):
    weighted_cells_graph.add_node(idx)
```

First of all let's join nodes which are remoat but have similar expression

```python
threshold_dist = numpy.mean(array_of_all_distance_probs)
threshold_gene = numpy.percentile(array_of_all_gene_exp_probs, 99.9)
expression_importance_coefficient = 100
for i in tqdm.tqdm(range(total_cell_amount)):
    for j in range(i + 1, total_cell_amount):
        dist_prob = array_of_all_distance_probs[i][j]
        expr_prob = array_of_all_gene_exp_probs[i][j]
        # Add the heavy edge if cells are remote and similar        
        if (dist_prob < threshold_dist) and (expr_prob > threshold_gene):
            weighted_cells_graph.add_edge(
                i, 
                j, 
                weight = expression_importance_coefficient * expr_prob
            )
```

```python
networkx.draw(
    weighted_cells_graph, 
    node_size=5, 
    pos=real_cell_poses, 
    edge_color="#808080", 
    width=.0007,
    node_color="black",
)
```

As a second step let's join nodes which are very close and have similar expression

```python
threshold_dist = numpy.percentile(threshold_dist, 99.93)
threshold_gene = numpy.percentile(array_of_all_gene_exp_probs, 99)
expression_importance_coefficient = 40
for i in tqdm.tqdm(range(total_cell_amount)):
    for j in range(i + 1, total_cell_amount):
        dist_prob = array_of_all_distance_probs[i][j]
        expr_prob = array_of_all_gene_exp_probs[i][j]
        # Add the normal edge if cells are very close and similar        
        if (dist_prob > threshold_dist) and (expr_prob > threshold_gene):
            weighted_cells_graph.add_edge(
                i, 
                j, 
                weight = expression_importance_coefficient * expr_prob
            )
```

```python
networkx.draw(
    weighted_cells_graph, 
    node_size=5, 
    pos=real_cell_poses, 
    edge_color="#808080", 
    width=.0007,
    node_color="black",
)
```

We can observe two types of edges on weights histogram

```python
weights = [weighted_cells_graph[u][v]['weight'] for u, v in weighted_cells_graph.edges]
```

```python
fig, ax = plt.subplots(figsize=(7, 4))
p1 = seaborn.histplot(data=[w for w in weights if w < 25], color="black", ax=ax)
ax.set(xlabel="Edge weight", ylabel="Count")
plt.show()
```

Now let's remove all small сomponents from the graph

```python
components = list(networkx.connected_components(weighted_cells_graph))
components_amount = len(components) 
biggest_component_size = max(list(map(len, components)))
print(f"There are {components_amount} components and bigger one has a size {biggest_component_size}")
```

Let's remove all isolated cells and small components, because we can't classify them precise

```python
biggest_component = [c for c in components if len(c) == biggest_component_size][0]
```

```python
filtered_weighted_cell_graph = weighted_cells_graph.subgraph(biggest_component)
```

```python
networkx.draw(
    filtered_weighted_cell_graph, 
    node_size=5, 
    pos=real_cell_poses, 
    edge_color="#808080", 
    width=.0005,
    node_color="black",
)
```

How this graph looks like if we remove layout:

```python
filtered_weights = [filtered_weighted_cell_graph[u][v]['weight'] for u, v in filtered_weighted_cell_graph.edges]
```

```python
_mask = [int(w > 12.5) + 1 for w in filtered_weights]
networkx.draw(
    filtered_weighted_cell_graph, 
    node_size=3, 
    pos=real_cell_poses, 
    edge_color=_mask,
    edge_cmap = plt.cm.Spectral,
    width=1 / 10 / numpy.array(_mask)**5 ,
    node_color="black",
)
```

### Clusterization
Let's find clusters of cells as a communities in the gene expression graph

```python
weights_based_clusters = list(
    networkx.community.louvain_communities(
        filtered_weighted_cell_graph,
        weight="weight"
    )
)
```

Remove all very small clusters 

```python
to_stay = []
filtered_clusters = []
for cluster in weights_based_clusters:
    if len(cluster) > 10:
        to_stay.extend(cluster)
        filtered_clusters.append(cluster)
```

```python
filtered_weighted_cell_graph = filtered_weighted_cell_graph.subgraph(to_stay)
```

```python
vertex_colors = dict()
for color, cluster in enumerate(filtered_clusters):
    for node in cluster:
        vertex_colors[node] = color
vertex_colors = [vertex_colors[node] for node in filtered_weighted_cell_graph.nodes]
print(len(set(vertex_colors)))
```

```python
palette = seaborn.color_palette("Spectral", n_colors=len(set(vertex_colors)))
legend_handles = [
    mpatches.Patch(
        color=palette[color], 
        label=f"Cell type {color}") 
    for color in set(vertex_colors)]

node_colors = [palette[color] for color in vertex_colors]

networkx.draw(
    filtered_weighted_cell_graph, 
    node_size=5, 
    pos=real_cell_poses, 
    node_color=node_colors, 
    edge_color="#808080", 
    width=.0005,
)

plt.legend(handles=legend_handles, loc="lower left")
plt.show()
```

### Save coordinates and cell types for further visualization

```python
final_vertex_id = []
final_vertex_xs = []
final_vertex_ys = []
final_cell_type = []
for v, c in zip(filtered_weighted_cell_graph.nodes, vertex_colors):
    final_vertex_id.append(ordered_cell_ids[v])
    final_vertex_xs.append(cells_coordinates[v][0])
    final_vertex_ys.append(cells_coordinates[v][1])
    final_cell_type.append(c)

clusterized_cell_dataset = pandas.DataFrame.from_dict(
    {
        "cell_id": final_vertex_id,
        "x": final_vertex_xs,
        "y": final_vertex_ys,
        "type": final_cell_type
    }
)
```

```python
fig, ax = plt.subplots(figsize=(4, 7))
p1 = seaborn.scatterplot(
    clusterized_cell_dataset, 
    x="x", 
    y="y", 
    hue="type", 
    palette=palette,
    s=10,
    linewidth=0,
    ax=ax,
)
ax.set(xlabel="x", ylabel="y")
plt.show()
```

```python
dataset_with_types_and_colors = pandas.DataFrame.from_dict(
    {
        "cell_id": final_vertex_id,
        "x": final_vertex_xs,
        "y": final_vertex_ys,
        "type": final_cell_type,
        "color": [palette.as_hex()[c] for c in vertex_colors] 
    }
)
```
